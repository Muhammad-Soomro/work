```python
# Group by 'VEHICLE_ID' and calculate time differences
def calculate_time_diffs(group):
    group['TIME_DIFFERENCE_BETWEEN_CONSECUTIVE_EVENTS'] = group['CREATED_AT'].diff()
    return group

def event_volumne_calculator(df):
    # Convert 'CREATED_AT' to datetime immediately
    df['CREATED_AT'] = pd.to_datetime(df['CREATED_AT'])

    # Sort by 'VEHICLE_ID' and 'CREATED_AT' only once
    df = df.sort_values(by=['VEHICLE_ID', 'CREATED_AT'], ascending=True)

    df_final = df.groupby('VEHICLE_ID').apply(calculate_time_diffs)

    # Reset index if needed
    df_final = df_final.reset_index(drop=True)
    
    return df_final

# Load data
file_path = './events_csv/Lane_Departure_Warning_EQA_from_2023_12_23_to_2024_01_02.csv'
df = pd.read_csv(file_path)
time_period = file_path.split('from_')[-1].split('.')[0]
df_final = event_volumne_calculator(df=df)
df_final.to_csv(f'/home/muhammadasadsoomro/Desktop/Work/LDW/event_vol_monitoring/event_volume_per_vehicle/{time_period}.csv' , index=False)
```
